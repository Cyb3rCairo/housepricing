```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
training_set = pd.read_csv("housepricing.csv")

X = training_set[
    [
        "bedrooms",
        "bathrooms",
        "sqft_living",
        "sqft_lot",
        "floors",
        "waterfront",
        "view",
        "condition",
        "sqft_above",
        "sqft_basement",
        "yr_built",
        "yr_renovated"
    ]
].values


y = training_set["price"].values
```

```python
def normalize(X):
    mu = X.mean(axis=0)
    sigma = X.std(axis=0)
    X_norm = (X - mu) / sigma
    return X_norm, mu, sigma
```

```python
X_norm, mu, sigma = normalize(X)
```

```python
def compute_cost(X, y, w, b):
    m = len(y)
    predictions = X @ w + b
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost
```

```python
def compute_gradient(X, y, w, b):
    m = len(y)
    predictions = X @ w + b
    error = predictions - y

    dj_dw = (1 / m) * (X.T @ error)
    dj_db = (1 / m) * np.sum(error)

    return dj_dw, dj_db
```

```python
def gradient_descent(X, y, alpha, iterations):
    n = X.shape[1]
    w = np.zeros(n)
    b = 0

    for i in range(iterations):
        dj_dw, dj_db = compute_gradient(X, y, w, b)

        w -= alpha * dj_dw
        b -= alpha * dj_db

        if i % 100 == 0:
            cost = compute_cost(X, y, w, b)
            print(f"Iter {i} | Cost {cost:.2f}")

    return w, b
```

```python
alpha = 0.01
iterations = 5000

w, b = gradient_descent(X_norm, y, alpha, iterations)

print("Pesos:", w)
print("Bias:", b)
```

```python
def predict(X, w, b):
    return X @ w + b
```

```python
y_pred = predict(X_norm, w, b)
```

```python
def rmse(y_true, y_pred):
    error = y_pred - y_true
    return np.sqrt(np.mean(error ** 2))

print("RMSE:", rmse(y, y_pred))
```
